```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("dataset\mail_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
mail_data = df.where((pd.notnull(df)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
mail_data.head()
```

```python
x = mail_data['Message']
y = mail_data['Category']
```

```python
z = mail_data.Message
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=3)
```

#feature extraction

```python
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
xtrain_features = feature_extraction.fit_transform(xtrain)
xtest_features = feature_extraction.transform(xtest)

ytrain = ytrain.astype('int')
ytest = ytest.astype('int')
```

```python
print(xtrain)
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain_features,ytrain)
```

```python
yprd = model.predict(xtrain_features)
accuracy = accuracy_score(ytrain,yprd)
```

```python
print("Accuracy on traning data : ",accuracy*100)
```

```python
yprd_test = model.predict(xtest_features)
accuracy_test = accuracy_score(ytest,yprd_test)
```

```python
print("Accuracy on test data : ",accuracy_test*100)
```

```python
input_mail = ["Congratulations! You’ve been selected as our lucky winner. Click here to claim your prize."]

input_data_feature = feature_extraction.transform(input_mail)

prediction = model.predict(input_data_feature)

if (prediction[0]==1):
  print("Ham Mail")
else:
  print("Spam Mail")
```
